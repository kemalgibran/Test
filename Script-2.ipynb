```python
import pandas as pd
from bs4 import  BeautifulSoup as bs
import requests
import re
```

```python
def collect_data(url):
    try: 
        r = requests.get(url).content
        soup = bs(r)
        column_1 = soup.find(class_='infobox').find_all('th')
        column_1 = [i.get_text(" ", strip=True) for i in column_1]
        attribute = column_1

        column_2 = soup.find(class_='infobox').find_all('td')
        column_2 = [i.get_text() for i in column_2]
        attribute_datas = column_2
        data = []

        for attribute_data in attribute_datas:
            if '\n' in attribute_data:
                y = attribute_data.replace('\n', 'dummy').replace('\xa0', ' ')
                y = y[1:].split('dummy')[:-1]
                data.append(y)
            elif attribute_data == '':
                print(url, 'error')
            else:
                y = attribute_data.replace('\xa0', ' ').replace('[1]', '')
                data.append(y)
        
        if len(attribute) != len(data):
            attribute = attribute[1:]

        title = soup.find('title').string.split(' -')[0]
        dict = {'Title': title}

        for i in range(len(data)):
            dict[attribute[i]] = data[i]

        return dict
    except:
        pass
```

```python
iurl = 'https://en.wikipedia.org/wiki/List_of_Walt_Disney_Pictures_films'
```

```python
r = requests.get(iurl).content
soup = bs(r)

links = soup.select('table.wikitable.sortable a')
list_link = ['https://en.wikipedia.org' + link['href'] for link in links]
list_link[50]
```

```python
movies = [collect_data(url) for url in list_link]
movies
```

```python
# movies
import json
with open('movies_data', 'w', encoding='utf-8') as fout:
    json.dump(movies, fout, ensure_ascii=False, indent=2)
```
